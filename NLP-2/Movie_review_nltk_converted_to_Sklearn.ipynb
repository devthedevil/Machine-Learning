# Data Cleaning

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
lemmatizer=WordNetLemmatizer()
w='better'
print(pos_tag([w]))
print(pos_tag(w))
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuation=list(string.punctuation)
stops.update(punctuation)
stops,string.punctuation
```

```python
def clean_review(words):
    out=[]
    for w in words:
        if w.lower() not in stops:
            #passing without lowering it because lowering can change pos of the word
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            out.append(clean_word.lower())
    return out
            
            
```

```python
doc=[(clean_review(document),category) for document,category in doc]
```

```python
doc[0]
```

# creating features

```python
train_doc=doc[:1500]
test_doc=doc[1500:]
```

```python
all_words=[]
for doc in train_doc:
    all_words+=doc[0]#appending the all first element of doc
all_words
```

```python
import nltk 
```

```python
freq=nltk.FreqDist(all_words)
common=freq.most_common(3000)
features=[i[0] for i in common]
```

```python
doc[1]
```

```python
def get_feature_dict(words):
    curr_features={}
    words_set=set(words)
    for w in features:
        curr_features[w]=w in words_set
    return curr_features
```

```python
get_feature_dict(train_doc[0][0])
```

```python
train_data=[(get_feature_dict(doc),cat) for doc ,cat in train_doc]
test_data=[(get_feature_dict(doc),cat) for doc ,cat in test_doc]
```

```python
train_data[0]
```

# Implementing naive bayes classifier from nltk

```python
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(train_data)
```

```python
nltk.classify.accuracy(classifier,test_data)
```

```python
classifier.show_most_informative_features(15)
```
