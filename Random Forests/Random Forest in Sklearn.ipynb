```python
# figure out useful columns for survival
# drop useless coluns
# adding new columns
# string changing into numeric
import pandas as pd
import numpy as np
titanic=pd.read_csv("titanic.csv") 
```

```python
titanic
```

```python
del titanic['Name']
del titanic["Fare"]
del titanic["PassengerId"]
del titanic["Cabin"]
del titanic["Ticket"]
titanic
```

```python
def f(s):
    if s=="male":
        return 1
    elif s=="female":
        return 2
titanic["Gender"]=titanic["Sex"].apply(f)
    
```

```python
def fa(s):
    if s=="Q":
        return 1
    elif s=="S":
        return 2
    else:
        return 3
titanic["embarked"]=titanic["Embarked"].apply(fa)
titanic.drop('Embarked',inplace=True,axis=1)
```

```python
titanic.drop('Sex',axis=1,inplace=True)

titanic
```

```python
titanic.rename(columns={"Gender":"Sex"},inplace=True)
titanic.describe()
```

```python
meanS=titanic[titanic.Survived==1].Age.mean()
meanS
meanNS=titanic[titanic.Survived==0].Age.mean()
meanNS
```

```python
titanic['age']=np.where(pd.isnull( titanic.Age) & titanic.Survived==1,meanS,titanic['Age'])
titanic.describe()
```

```python
titanic.age.fillna(meanNS,inplace=True)
titanic.describe()
```

```python
del titanic["Age"]
titanic.rename(columns={"age":"Age"},inplace=True)
print(titanic)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(titanic.iloc[:,1:],titanic.iloc[:,0])
x_train.isnull().sum()
x_test.isnull().sum()
y_train.isnull().sum()
y_test.isnull().sum()
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
clf.fit(x_train,y_train)

y_pred=clf.predict(x_test)
clf.score(x_train,y_train),clf.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(max_depth=6,n_estimators =100,random_state=0,
                           criterion='entropy',min_samples_split=10,
                           min_samples_leaf=4,
                           ccp_alpha=2)
clf.fit(x_train,y_train)

y_pred=clf.predict(x_test)
clf.score(x_train,y_train),clf.score(x_test,y_test)
```
