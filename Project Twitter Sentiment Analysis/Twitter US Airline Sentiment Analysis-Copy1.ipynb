# Data Cleaning

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops=set(stopwords.words('english'))
punctuations=list(string.punctuation)
# stops.update(punctuations)
# stops
# pos_tag("he")
```

```python
def clean_review(words):
    out=[]
    for w in words:
        if w.lower() not in stops:
            #passing without lowering it because lowering can change pos of the word
            if w.isdigit():# or  len(w)<3:
                continue
            if w[0]=="'":
                w=w[1:]
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            out.append(clean_word.lower())
    return out
            
            
```

```python
doc=[(clean_review(document),category) for document,category in doc]
doc_test=[(clean_review(document)) for document in doc_test]
```

```python

categories=[category for document,category in doc ]
categories
```

```python
text_doc=[" ".join(document) for document,cat in doc]
text_doc_test=[" ".join(document) for document in doc_test]
```

```python
text_doc[0]
```

```python
x_train=text_doc
y_train=categories
x_test=text_doc_test
```

```python
count_vec=TfidfVectorizer(max_features=2000,ngram_range=(1,3))#,max_df=0.8,min_df=0.1)
x_train_features=count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
x_test_features=count_vec.transform(x_test)
x_test_features.todense
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(x_train_features,y_train)
```

```python
y_pred=svc.predict(x_test_features)
```
